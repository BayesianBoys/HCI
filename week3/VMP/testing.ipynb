```python
# very simple function
def plot_scatter(df, x, y, size = None, hue = None): 
    
    sns.scatterplot(x = x, y = y, 
                    size = size, hue = hue, 
                    data = df)
```

```python
plot_scatter(g_15, x = "GDP_capita", y = "life_expectancy", size = "Happiness Rank")
```
